```python
#exemplo de orientação a objeto.

class carro():
    def __init__(self, carrro, modelo):
        self.carrro = carrro
        self.modelo = modelo
        
    def carro(self):
        print("Carro : ", self.carrro)
        print("modelo : ",self.modelo)
        
carros = carro('corsa','g5')
carros.carro()
```
