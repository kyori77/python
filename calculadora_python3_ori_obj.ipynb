```python
#'''calculadora em python orientado a objeto'''

class oper():
  def __init__(self, num1, num2):
    self.num1 = num1
    self.num2 = num2

  def soma(self):
    return self.num1 + self.num2

  def sub(self):
    return self.num1 - self.num2

  def div(self):
    return self.num1 / self.num2

  def mult(self):
    return self.num1 * self.num2

num1 = float(input("informe o primeiro numero : "))
num2 = float(input("informe o primeiro numero : "))
op = input("Informe operação +,-,*,/ ")
result = oper(num1, num2)

if op == '+':
  print(result.soma())
elif op == '-':
  print(result.sub())
elif op == '/':
  print(result.div())
else:
  print(result.mult())
```
