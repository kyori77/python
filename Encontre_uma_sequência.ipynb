```python
def count_substring(string, sub_string):

  cont = 0

  for letra in range(0, (len(string)-len(sub_string)+1)):
    if (string[letra:(letra+len(sub_string))]) == sub_string:
      cont+=1
  
  return cont

if __name__ == '__main__':
  string = input().strip()
  sub_string = input().strip()
    
  count = count_substring(string, sub_string)
  print(count)
```
