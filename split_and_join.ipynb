```python
def split_and_join(line):
  line = line.split()#comando split remove os espaços da string ou qual seja a condição no ()
  line = "-".join(line)#comando join inclui algo em uma string
  return line

if __name__ == '__main__':
  line = input()
  result = split_and_join(line)
  print(result)
```
