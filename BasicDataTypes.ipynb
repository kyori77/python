```python
lista = []
if __name__ == '__main__':
  n = int(input())

  for _ in range(n):
    name, *line = input().split()
    scores = list(map(int, line))

    if (name == "append"):
      lista.append(scores[0])
    elif (name == "insert"):
      lista.insert(scores[0], scores[1])
    elif (name == "reverse"):
      lista.reverse()
    elif (name == "sort"):
      lista.sort()
    elif(name == "remove"):
      lista.remove(scores[0])
    elif(name == "pop"):
      lista.pop()
    elif(name == "print"):
      print(lista)
```
