```python
def swap_case(s):
  palavra = ""
  for letra in s:
    if letra == letra.lower():
      letra = letra.upper()
      palavra += letra
    else:
      letra = letra.lower()
      palavra += letra
  return palavra

if __name__ == '__main__':
  s = str(input())
  result = swap_case(s)
  print(result)
```
