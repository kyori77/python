```python
lsName = ["Harry","Berry","Tina","Akriti","Harsh"]
lsScore = [37.21,37.21,37.2,41,39]
nam = []

if __name__ == '__main__':

  for _ in range(int(input())):
    name = input()
    score = float(input())
    lsName.append(name)
    lsScore.append(score)
    
  maior = list(sorted(set(lsScore)))
  maior = (maior[1])

  ls = len(lsScore)
  lista = 0

  while lista < ls:
    if(lsScore[lista] == maior):
      nam.append(lsName[lista])
    lista+=1

  nam = sorted(nam)

  for nome in nam:
    print(nome)
```
