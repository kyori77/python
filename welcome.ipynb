```python

altura, largura = input().split()
largura = int(largura) - 3
altura = int(altura)
c = '.|.'

#Top Cone
for i in range(int(altura/2)):
  print((c*i).rjust(int(largura/2),'-')+c+(c*i).ljust(int(largura/2),'-'))

print('-'*(int((largura-4)/2))+"WELCOME"+'-'*(int((largura-4)/2)))
reverso = (int(altura/2))-1

while reverso >= 0:
  print((c*reverso).rjust(int(largura/2),'-')+c+(c*reverso).ljust(int(largura/2),'-'))
  reverso-=1
```
