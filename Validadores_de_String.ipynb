```python
if __name__ == '__main__':
  s = input()

  alfanum = False
  alfbet = False
  dig = False
  low = False
  up = False

  for x in range(0, len(s)):
    if s[x].isalnum() == True:
      alfanum = True
    if s[x].isalpha() == True:
      alfbet = True
    if s[x].isdigit() == True:
      dig = True
    if s[x].islower() == True:
      low = True
    if s[x].isupper() == True:
      up = True

print(alfanum)
print(alfbet)
print(dig)
print(low)
print(up)

#print(s.isalnum())#Este método verifica se todos os caracteres de uma string são alfanuméricos (az, AZ e 0-9).
#print(s.isalpha())#Este método verifica se todos os caracteres de uma string são alfabéticos (az e AZ).
#print(s.isdigit())#Este método verifica se todos os caracteres de uma string são dígitos (0-9).
#print(s.islower())#Este método verifica se todos os caracteres de uma string são caracteres minúsculos (az).
#print(s.isupper())#Este método verifica se todos os caracteres de uma string são caracteres maiúsculos (AZ).
```
